```python
# importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
# for visualisation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Importing the datasets
college_df = pd.read_csv("final_college_df.csv", index_col=0)
college_fr = pd.read_csv("fr_college_df.csv", index_col=0)
college_so = pd.read_csv("so_college_df.csv", index_col=0)
college_jr = pd.read_csv("jr_college_df.csv", index_col=0)
college_sr = pd.read_csv("sr_college_df.csv", index_col=0)
```

```python
#college_train = college_df[college_df['year'] <= 2018]
#college_train
```

```python
college_fr_train = college_fr[college_fr['year'] <= 2018]
college_so_train = college_so[college_so['year'] <= 2018]
college_sr_train = college_sr[college_sr['year'] <= 2018]
college_jr_train = college_jr[college_jr['year'] <= 2018]
```

```python
college_fr_pred = college_fr[college_fr['year'] > 2018]
college_so_pred = college_so[college_so['year'] > 2018]
college_sr_pred = college_sr[college_sr['year'] > 2018]
college_jr_pred = college_jr[college_jr['year'] > 2018]
college_fr_pred
```

```python
college_fr_train = college_fr_train.drop('yr',axis=1)
college_so_train = college_so_train.drop('yr',axis=1)
college_sr_train = college_sr_train.drop('yr',axis=1)
college_jr_train = college_jr_train.drop('yr',axis=1)
college_fr_pred = college_fr_pred.drop('yr',axis=1)
college_so_pred = college_so_pred.drop('yr',axis=1)
college_sr_pred = college_sr_pred.drop('yr',axis=1)
college_jr_pred = college_jr_pred.drop('yr',axis=1)
```

```python
#Freshman training
x_train = college_fr_train.iloc[:,2:-1].values
y_train = college_fr_train.iloc[:,-1].values
y_train = y_train.reshape(-1, 1)
```

```python
x_pred = college_fr_pred.iloc[:,2:-1].values
y_pred = college_fr_pred.iloc[:,-1].values
y_pred = y_pred.reshape(-1, 1)
```

```python
# create an empty linear regression model like below and give it a good variable name
fr_model = LinearRegression()

# to create the model, we use fit(x,y)
fr_model.fit(x_train,y_train)

predictions = fr_model.predict(x_pred)
```

```python
college_fr_pred['test_score'] = predictions
```

```python
predictions = fr_model.predict(x_train)
college_fr_train['test_score'] = predictions
```

```python
college_fr_pred.to_csv('fr_scores.csv')
```

```python
college_fr_train['diff'] = college_fr_train['test_score'] - college_fr_train['rating']
college_fr_train.to_csv('analyse_fr_scores.csv')
```
