```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
college_df = pd.read_csv("college_df_with_scores.csv", index_col=0)
college_df
```

```python
ratings_df = pd.read_csv('player_with_only_rating.csv', index_col=0)
rating_df = pd.read_csv('player_with_only_rating.csv')
```

```python
ratings = ratings_df.to_dict()['total_score']
```

```python
ratings
```

```python
college_names = college_df['player_name'].values
rating_names = rating_df['player_name'].values
```

```python
scores = []
for name in college_names:
    if name in rating_names:
        scores.append(ratings[name])
    else:
        scores.append(0)

college_df['my_score'] = scores
```

```python
college_df.to_csv('college_df_with_scores.csv')
```
