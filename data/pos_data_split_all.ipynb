```python
yr = college_df['yr'].values
seasons = []
for year in yr:
    if year == 'Fr':
        seasons.append(4)
    elif year == 'So':
        seasons.append(3)
    elif year == 'Jr':
        seasons.append(2)
    elif year == 'Sr':
        seasons.append(1)
    else:
        seasons.append(0)
college_df['season'] = seasons
```
